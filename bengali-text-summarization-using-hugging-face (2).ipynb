```python
def semantic_dedup(sentences,embeding,threshold=0.85):
    keep = []
    keep_emb = []
    for i,sent in enumerate(sentences):
        if not keep:
            keep.append(sent)
            keep_emb.append(embeding[i])
            continue
        sim = cosine_similarity([embeding[i]],keep_emb).max()
        if sim < threshold:
            keep.append(sent)
            keep_emb.append(embeding[i])

    return keep
            
            
    
```

```python
def clean_articles(text):
    sentence = str(text).split("|")
    embedding = model.encode(sentence)
    clean = semantic_dedup(sentence,embedding,0.85)
    return "|".join(clean)

    
```

```python
import re

def full_clean_text(text):
    text = str(text)
    # remove unicode noise
    text = text.replace("\u200b",'')
    text = text.replace("\ufeff",'')
    text = text.replace("\u200c",'')
    text = text.replace("\u200d",'')
    # remove html tages
    text = re.sub(r"<.*?>","",text)
    # remove urls
    text = re.sub(r"http\S+|www\S+","",text)
    # normalize puctuation
    text = re.sub(r"\|+","|",text)
    text = re.sub(r"\?+","?",text)
    text = re.sub(r"!+","!",text)

    text = re.sub(r"\s+"," ",text).strip()
    return text
    

    
```

```python
def convert_example_to_feature(example_batch,):
    
    inputs = ["summarize: " + text for text in example_batch['clean_text'] ]
    
    input_encoding = tokenizer(inputs,max_length=1024,padding= "max_length",truncation=True)
    
    target_encoding = tokenizer(text_target= example_batch['summary'],max_length=128,padding="max_length",truncation=True)
    labels = target_encoding["input_ids"]
    labels = [
        [(token if token!=tokenizer.pad_token_id else -100 ) for token in label]
        for label in labels
    ]
    

    return {
         "input_ids": input_encoding["input_ids"],
         "attention_mask": input_encoding["attention_mask"],
         "labels": labels
     }
```

```python
# for huggingface dataset instead of apply we need to use map
tokenize_dataset = dataset.map(convert_example_to_feature,batched=True, remove_columns=['summary', 'clean_text', '__index_level_0__']) 
```

```python
# Function for batch 
def generate_batch_size(list_of_elements, batch_size):
    for i in range(0,len(list_of_elements), batch_size):
        batch = list_of_elements[i:i+batch_size]
        yield batch # এখুনি ব্যাচ রিটার্ন করে, কিন্তু ফাংশন শেষ করে না
```

```python
def calculate_metric_on_test_ds(
    dataset,
    metric,
    model,
    tokenizer,
    batch_size,
    column_text="article",
    column_summary="highlights"
):
    articles_batch = list(
        generate_batch_size(dataset[column_text], 
                            batch_size)
    )
    targets_batch  = list(
        generate_batch_size(dataset[column_summary], 
                            batch_size)
    )
    model.eval()
    for articles,targets in tqdm(
        zip(articles_batch,targets_batch),
        total= len(articles_batch)
    ):
        inputs = tokenizer(
            articles,
            max_length=1024,
            padding= "max_length",
            truncation=True,
            return_tensors="pt"
        ).to(device) 
        with torch.no_grad():
            summarise = model.generate(
                input_ids = inputs["input_ids"],
                attention_mask= inputs["attention_mask"],
                length_penalty = 0.8,
                num_beams= 8,
                max_length=64,
                min_length =10
                
            ).to(device)
        decode_summarise=[
            tokenizer.decode(
                s,
                skip_special_tokens=True,
                clean_up_tokenization_spaces=True
                
            )
            for s in summarise
        ]
        decode_summarise = [
            summary.strip()
            for summary in decode_summarise
        ]
        targets = [
            str(t).strip()
            for t in targets
        ]
        metric.add_batch(
            predictions=decode_summarise, 
            references=targets
        )
    score = metric.compute()
    return score
    
        
        
      
        
        
        
        
```

```python
score_1 = calculate_metric_on_test_ds(
    dataset['train'][0:100], rouge, trainer.model, tokenizer, batch_size = 2, column_text = 'clean_text', column_summary= 'summary'
) 
```

```python
score_2 = calculate_metric_on_test_ds(
    dataset['test'][0:100], rouge, trainer.model, tokenizer, batch_size = 2, column_text = 'clean_text', column_summary= 'summary'
) 
```
